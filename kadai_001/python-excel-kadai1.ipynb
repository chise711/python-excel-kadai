```python
import openpyxl

from datetime import datetime

workbook = openpyxl.Workbook()

# シートの選択
sheet = workbook.active

# セルに値を設定
sheet['B2'].value = '請求書'
sheet['B4'].value = '株式会社ABC'
sheet['B5'].value = '〒101-0022 東京都千代田区神田練塀町300'
sheet['B6'].value = 'TEL:03-1234-5678 FAX:03-1234-5678'
sheet['B7'].value = '担当者名:鈴木一郎 様'

Product_Name = ['商品名','商品A','商品B']
Count = ['数量',2,1]
Unit_Price = ['単価',10000,15000]
Amount_Price = ['金額',20000,15000]

for i in range(0,len(Product_Name)):

  #B列にリストを書き込み
  sheet.cell(i+10,2,value = Product_Name[i])
  #C列にリストを書き込み
  sheet.cell(i+10,3,value = Count[i])
  #D列にリストを書き込み
  sheet.cell(i+10,4,value = Unit_Price[i])
  #E列にリストを書き込み
  sheet.cell(i+10,5,value = Amount_Price[i])

sheet['B15'].value = '合計'
sheet['B16'].value = '消費税'
sheet['B17'].value = '税込合計'
sheet['E13'] = 35000
sheet['E15'] = 35000
sheet['E16'] = 3500
sheet['E17'] = 38500
sheet['F4'].value = 'No.'
sheet['F5'].value = '日付'
sheet['G4'].value = '0001'

today = datetime.today()
sheet['G5'] = today.strftime('%Y/%m/%d')

sheet.column_dimensions['B'].width = 10
sheet.column_dimensions['C'].width = 10
sheet.column_dimensions['D'].width = 13
sheet.column_dimensions['E'].width = 13

# 現在の日付（年月日）を取得して、変数todayに代入する
today = datetime.today()
workbook.save(f"請求書_{today.strftime('%Y%m%d')}.xlsx")
```
